```python
import re

patron = r"^(.*?)(\(\d+\))$"
lista_total = []
for dict in media_per_year:
    for category in ['Anime & Manga', 'Books & Literature', 'Cartoons & Comics & Graphic Novels', 'Celebrities & Real People', 'Music & Bands', 'Other Media', 'Theater', 'TV Shows', 'Video Games']:
        for title in dict[f'{category}']:
            lista_aux = [dict['year'], f'{category}']
            match = re.match(patron, title)
            if match:
                texto, numero = match.groups()
                lista_aux.append(texto.strip())
                lista_aux.append(int(numero.replace('(', '').replace(')', '')))
            lista_total.append(tuple(lista_aux))
    
```
